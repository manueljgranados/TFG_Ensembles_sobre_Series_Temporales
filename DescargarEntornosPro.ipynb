```python
"""
Descarga de escenarios:
Escenario pequeno (ventana:20, horizonte:1):
S1 --> Del 01/03/2025 al 21/03/2025.
S2 --> Del 01/06/2025 al 21/06/2025.
S3 --> Del 01/09/2025 al 21/09/2025.

Escenario mediano (ventana:60, horizonte:5):
M1 --> Del 15/01/2025 al 16/03/2025.
M2 --> Del 15/04/2025 al 14/06/2025.
M3 --> Del 16/07/2025 al 14/09/2025.

Escenario grande (ventana:120, horizonte:20):
L1 --> Del 02/01/2025 al 02/05/2025.
L2 --> Del 02/03/2025 al 30/06/2025.
L3 --> Del 02/05/2025 al 30/08/2025.
"""

import yfinance as yf

tickers = [
    "^GSPC","^IXIC","^DJI","^RUT",
    "^FTSE","^GDAXI","^FCHI","^125904-USD-STRD","^IBEX",
    "^N225","^HSI","000001.SS","^KS11","^BSESN",
    "^GSPTSE","^BVSP","^MXX","^MERV",
    "^AXJO","^NZ50",
    "ES=F","NQ=F","YM=F","ZT=F","^VIX"
]

# Descarga escenario S1 (21 dias laborables)
data = yf.download(
    tickers,
    start="2025-02-28",
    end="2025-03-31"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_S1.csv")

# Descarga escenario S2 (21 dias laborables)
data = yf.download(
    tickers,
    start="2025-05-23",
    end="2025-06-22"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_S2.csv")

# Descarga escenario S3 (21 dias laborables)
data = yf.download(
    tickers,
    start="2025-07-23",
    end="2025-08-21"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_S3.csv")



# Descarga escenario M1 (65 dias laborables)
data = yf.download(
    tickers,
    start="2025-01-02",
    end="2025-04-03"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_M1.csv")

# Descarga escenario M2 (65 dias laborables)
data = yf.download(
    tickers,
    start="2025-03-24",
    end="2025-06-23"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_M2.csv")

# Descarga escenario M3 (65 dias laborables)
data = yf.download(
    tickers,
    start="2025-06-12",
    end="2025-09-11"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_M3.csv")


# Descarga escenario L1 (140 dias laborables)
data = yf.download(
    tickers,
    start="2025-01-02",
    end="2025-07-17"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_L1.csv")

# Descarga escenario L2 (140 dias laborables)
data = yf.download(
    tickers,
    start="2025-02-05",
    end="2025-08-20"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_L2.csv")

# Descarga escenario L3 (140 dias laborables)
data = yf.download(
    tickers,
    start="2025-02-27",
    end="2025-09-11"
)

cierres = data["Close"]  # DataFrame con cada ticker como columna
cierres.to_csv("cierres_diarios_L3.csv")
```
